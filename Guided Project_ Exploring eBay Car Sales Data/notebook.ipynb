The aim of this project is to clean the data and analyze the included used car listings.

The data consists of 20 columns and 50000 rows with most of the data types are objects. Dates column are also recognized as object. If we are looking at the data more thouroughly, we can see that `price` and `odometer` columns are read as object, that is because the values have additional string in it.

We are now going to replace the naming of the columns, which use camelCase, with snakeCase

```python
autos.columns
```

```python
autos.rename({'dateCrawled':'ad_crawled',
              'offerType':'offer_type',
              'vehicleType':'vehicle_type',
              'yearOfRegistration':'registration_year',
              'monthOfRegistration':'registration_month',
              'powerPS':'power_ps',
              'fuelType':'fuel_type',
              'notRepairedDamage':'unrepaired_damage',
              'dateCreated':'ad_created',
              'nrOfPictures':'num_of_pics',
              'postalCode':'postal_code',
              'lastSeen':'last_seen'}, axis=1, inplace=True)
```

```python
autos.head()
```

This is done for easier read on the columns and standardization.

```python
autos.describe(include='all')
```

There are several actions that should be done after looking at the description:
* Any columns that have mostly one value that are candidates to be dropped
* Any column that need more investigation
* Any examples of numeric data stored as text that needs to be cleaned

```python
autos['price'] = autos['price'].str.replace('$', '').str.replace(',', '').astype(int)
autos['odometer'] = autos['odometer'].str.replace('km', '').str.replace(',', '').astype(int)
```

```python
autos.rename({'odometer':'odometer_km'}, axis=1, inplace=True)
```

There are outliers starting from 999990. We are going to remove this

```python
autos = autos[~(autos['price'] > 999989)]
```

```python
autos = autos[autos['registration_year'].between(1900,2016)]
autos['registration_year'].value_counts(normalize=True).head(10)
```

```python
autos["brand"].value_counts(normalize=True)
```

```python
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > .05].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    brand_only = autos[autos['brand'] == brand]
    mean_price = brand_only['price'].mean()
    brand_mean_prices[brand] = int(mean_price)
    
brand_mean_prices    
```
